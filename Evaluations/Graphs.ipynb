```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Multiple Bar Plot

```python
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
 
# set height of bar
selenium = [12, 30, 1, 8, 22]
scrapy = [28, 6, 16, 5, 10]
reaper = [29, 3, 24, 25, 17]
 
# Set position of bar on X axis
br1 = np.arange(len(selenium))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
 
# Make the plot
plt.bar(br1, selenium, color ='r', width = barWidth,
        edgecolor ='grey', label ='selenium')
plt.bar(br2, scrapy, color ='g', width = barWidth,
        edgecolor ='grey', label ='scrapy')
plt.bar(br3, reaper, color ='b', width = barWidth,
        edgecolor ='grey', label ='reaper')
 
# Adding Xticks
plt.xlabel('Scraper', fontweight ='bold', fontsize = 15)
plt.ylabel('Metric numbers', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(selenium))],
        ['Time', 'Lines of code', 'Library functions', 'Scalability', 'Learning Curve'])
 
plt.legend()
plt.show()
```
